```python
import matplotlib.pyplot as plt
import numpy as np

# Número de términos a calcular (puedes aumentar si quieres)
n = 20

# Generamos los índices de 1 a n
x = np.arange(1, n+1)

# Calculamos los términos parciales H_n = sum(1/k)
H = np.cumsum(1/x)

# Graficamos
plt.figure(figsize=(10,6))
plt.plot(x, H, label='Suma parcial de 1/n', color='blue')
plt.xlabel('n')
plt.ylabel('H_n')
plt.title('Crecimiento de la serie armónica')
plt.grid(True)
plt.legend()
plt.show()
```
